```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
single_laplacian = pd.read_csv('test_metrics_single_laplacian_gudhi_medical.csv')
dual_abstract = pd.read_csv('test_metrics_dual_lap_abstract_gudhi_medical.csv')
dual_vr = pd.read_csv('test_metrics_dual_lap_vr_gudhi_medical.csv')
```

```python
single_laplacian
dual_abstract
dual_vr
```

```python
single_laplacian['Dataset'] = "Single Laplacian"
dual_abstract['Dataset'] = "Dual Laplacian + Abstract"
dual_vr['Dataset'] = "Dual Laplacian + VR"
combined_data = pd.concat([single_laplacian, dual_abstract, dual_vr])
metric_to_plot = "Recall"
```

```python
plt.figure(figsize=(10,6))
for labels, group in combined_data.groupby('Dataset'):
    plt.plot(group['Epoch'], group[metric_to_plot], label=labels)

plt.title(f"Epoch vs {metric_to_plot}", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel(metric_to_plot, fontsize=12)
plt.legend(title="Dataset", fontsize=10)
plt.grid(alpha=0.5)
plt.show()
```
